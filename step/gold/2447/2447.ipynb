```python
# 공백이 될 x,y 좌표를 얻어내기
space = []
for x in range(0,9):
    for y in range(0,9):
        for i in range(2):
            start = 3 ** i
            end = start * 2
            new_x = x % (3 ** (i+1))
            new_y = y % (3 ** (i+1))
            if new_x in range(start, end) and new_y in range(start, end):
                space.append((x,y))
                break
print(space)
```

```python
import sys
space = []
for x in range(0,9):
    line = []
    for y in range(0,9):
        isStar = True
        for i in range(2):
            start = 3 ** i
            end = start * 2
            new_x = x % (3 ** (i+1))
            new_y = y % (3 ** (i+1))
            if new_x in range(start, end) and new_y in range(start, end):
                isStar = False
                break
        if isStar:
            line.append('*')
        else:
            line.append(' ')
    space.append(line)
output = '\n'.join(
    "".join(map(str, row)) for row in space
)
sys.stdout.write(output)
```
